```python
import numpy as np
```

```python
N = 10
```

```python
A = np.random.rand(N,N) + np.random.rand(N,N)*1j
B = np.random.rand(N) +  + np.random.rand(N)*1j
```

```python
A
```

```python
A_inv = np.linalg.inv(A)
```

```python
x = np.matmul(A_inv, B)
print(x)
```
